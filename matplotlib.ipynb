```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])

plt.plot(y, linestyle = 'dotted', color='green', marker='o')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.show()
```

```python
x = np.array([1,2,3,4])
y1 = np.array([2,4,6,8])
y2 = np.array([1,3,5,7])

plt.plot(x,y1, label='Line1', color='red', marker='s', linestyle='dotted')
plt.plot(x,y2, label='Line2', color='green', marker='o', linestyle='dashed')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.legend()
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y, color='red')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Bar Plot")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,8,1,10,12])

plt.scatter(x,y, color='purple', s=100)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
data = np.array([3,4,5,6,7,8,9,10])

plt.hist(data, color='skyblue', edgecolor='black', bins=5)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Histogram Plot")
plt.show()
```

```python
x = np.array([1,2,3,4])
y1 = np.array([1,4,9,16])
y2 = np.array([1,3,5,7])

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.plot(x,y1, color='red')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")

plt.subplot(1,2,2)
plt.bar(x,y2, color='green')
plt.tight_layout()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")

plt.show()
```

```python
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])

plt.plot(x,y,color='green')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.savefig('myplot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot:
fig, ax = plt.subplots()

ax.hexbin(x, y, gridsize=20)

ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.show()
```
